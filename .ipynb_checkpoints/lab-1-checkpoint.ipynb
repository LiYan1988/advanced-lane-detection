```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from glob import glob

%matplotlib inline
```

```python
nx, ny = 9, 6
img_list = []
corner_list = []
for fname in glob('camera_cal/calibration*.jpg'):
    img = cv2.imread(fname)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ret, corner = cv2.findChessboardCorners(img, (nx, ny))
    if ret:
        cv2.drawChessboardCorners(img, (nx, ny), corner, ret)
        img_list.append(img)
        corner_list.append(corner)
        print(fname)
```

```python
objpts = np.zeros((nx * ny, 3), np.float32)
objpts[:, :2] = np.mgrid[0:nx, 0:ny].T.reshape(-1, 2)
objpts = np.tile(objpts, (len(img_list), 1, 1))
objpts = list(objpts)
```

```python
_, cameraMatrix, distCoeffs, _, _ = cv2.calibrateCamera(objpts, corner_list, img_list[0].shape[1::-1], None, None)
```

```python
for fname in glob('camera_cal/calibration*.jpg'):
    img = cv2.imread(fname)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_undist = cv2.undistort(img, cameraMatrix, distCoeffs, None, cameraMatrix)
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.subplot(1, 2, 2)
    plt.imshow(img_undist)
    plt.show()
```

```python
import pickle
with open('camera_parameters.p', 'wb') as pfile:
    pickle.dump(pfile, {'cameraMatrix': cameraMatrix, 'distCoeffs': cameraMatrix})
```
